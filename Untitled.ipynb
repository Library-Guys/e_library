```python
from asgiref.sync import sync_to_async
from books.models import Review

@sync_to_async
def get_books():
    return list(Pdf_Info.objects.all().values())

books = await get_books()
df_books = pd.DataFrame(books)
df_books.head()
# books = df_books.rename(columns={'id':'book_id'}, inplace = True)
# books
```

```python
from asgiref.sync import sync_to_async
from auth_users.models import Profile

@sync_to_async
def get_users():
    return list(Profile.objects.all().values())

users = await get_users()
df_users = pd.DataFrame(users)
df_users.head()
```

```python
@sync_to_async
def get_reviews():
    return list(Review.objects.all().values())

reviews = await get_reviews()
df_reviews = pd.DataFrame(reviews)
df_reviews.head(50)
```

```python
## Popularity based Recommendation System
```

```python
ratings_with_name = df_reviews.merge(df_books, on = 'id')
```

```python

num_rating_df = ratings_with_name.groupby('title')['rating'].count().reset_index()
num_rating_df.rename(columns={'rating':'num_ratings'}, inplace = True)
num_rating_df
```

```python
#Average Rating
avg_rating_df = ratings_with_name.groupby('title')['rating'].mean().reset_index()
avg_rating_df.rename(columns={'rating':'avg_rating'}, inplace = True)
avg_rating_df 
```

```python
#merge average rating and number rating
popular_df = num_rating_df.merge(avg_rating_df, on = 'title')
popular_df
```

```python
#shows popular book whose num rating>3
popular_df = popular_df[popular_df['num_ratings'] >=2].sort_values('avg_rating',ascending =False)
popular_df
```

```python
popular_df.merge(df_books, on='title')
```

```python
popular_df = popular_df.merge(df_books, on='title').drop_duplicates('title')[['title', 'author', 'image', 'num_ratings', 'avg_rating']]
```
